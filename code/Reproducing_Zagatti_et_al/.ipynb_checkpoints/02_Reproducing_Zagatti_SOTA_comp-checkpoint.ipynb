# Importing libs

```python
#import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.io.arff import loadarff
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
#import copy
#import random
#import scikit_posthocs as sp
#import scipy.stats as stats
#from scipy.stats import normaltest
#from scipy.stats import friedmanchisquare
#import pickle
#import seaborn as sns
#import Orange

from metaprep.metaprep import cleaner
```

```python
# Loading the 300 datasets from ICPR 2010
datasets = []
for i in range(1,301):
    raw_data = loadarff(f'../../data/S1/D{i}-trn.arff')
    df_data = pd.DataFrame(raw_data[0])
    datasets.append(df_data)
    
# Transforming class labels into ones and zeros.
for ds in datasets:
    ds['class'] = preprocessing.LabelEncoder().fit_transform(ds['class'])
```

```python
# Example of Metaprep execution (Zagatti et al. 2021):
mp_cleaner = cleaner()
mp_cleaner.set_dataframe(datasets[254])
candidate_pipelines = mp_cleaner.meta_prep(target='class', verbose=1)
print(candidate_pipelines[0][0])
```

```python
# Applying Metaprep to our datasets and saving the recommended ST in y_pred.
y_pred = []
for ds in datasets:
    mp_cleaner = cleaner()
    mp_cleaner.set_dataframe(ds)
    candidate_pipelines = mp_cleaner.meta_prep(target='class', verbose=1)
    recom_st = candidate_pipelines[0][0]
    if recom_st == 'normalizer':
        y_pred.append('NOR')
    elif recom_st == 'standard_scaler':
        y_pred.append('SS')
    elif recom_st == 'minmax':
        y_pred.append('MM')
    else:
        y_pred.append('NS')
```

```python
y_pred
```

# Loading ST performances

```python
performs_6_STs = pd.read_csv('../../results/csv_tabs/results_ST_perfs.csv')
```

```python
performs_NOR = pd.read_csv('results_NOR_perfs.csv').iloc[:,1:]
```

```python
performs_NOR.Model.unique()
```

```python
performs_all = pd.concat([performs_6_STs, performs_NOR], ignore_index='True')
performs_all
```

```python
metaprep_clf_perfs = {}
for model in performs_all.Model.unique():
    print(model, end=' ')
    metaprep_clf_perfs[model] = []
    for ds in range(1,301):
        p_mod = performs_all[performs_all['Model'] == model]
        p_mod_ds = p_mod[p_mod['Dataset'] == ds]
        p_mod_ds_st = p_mod_ds[p_mod_ds['Scaling technique'] == y_pred[ds-1]]
        metaprep_clf_perfs[model].append(p_mod_ds_st.f1_mean.values[0])
        print('.', end='')
    
```
