# Getting and arranging training pipelines

```python
import pandas as pd
import time
import numpy as np
from sklearn import metrics
from utils import separate
import sklearn.model_selection
from sklearn.ensemble import RandomForestClassifier
from metaprep import cleaner
import pickle

pipelines = pd.read_csv('pipelines/pipelines.csv')
pipelines.head()
```

```python
pipeline = []
dic = {}

#Creating pipelines list
df_dict = dict(
    list(
        pipelines.groupby(pipelines.index)
    )
)

for k, v in df_dict.items():         
    check = v.columns[(v == 1).any()]
    pipeline.append(list(check))
    
print(*pipeline, sep='\n')
```

# Function to separate datasets in specific cases

```python
def choose_case(data, target):
    case = []
    df = data.get_dataframe()
    total_null = df.isna().sum().sum()
    num_columns, cat_columns, _, _, _ = separate(df, target)
    if(cat_columns is None):
        cat_columns = 0;
    else:
        cat_null = cat_columns.isna().sum().sum()
        cat_columns = 1

    if(num_columns is None):
        num_columns = 0
    else:
        num_null = num_columns.isna().sum().sum()
        num_columns = 1

    if((cat_columns == 0) and (total_null == 0)): 
        case1 = pd.read_csv('pipelines/case1.csv')
        df_dict = dict(
            list(
                case1.groupby(case1.index)
            )
        )

        for k, v in df_dict.items():         
            check = v.columns[(v == 1).any()]
            case.append(list(check))
            
    elif((cat_columns == 0) and (num_null > 0)): 
        case2 = pd.read_csv('pipelines/case2.csv')
        df_dict = dict(
            list(
                case2.groupby(case2.index)
            )
        )

        for k, v in df_dict.items():         
            check = v.columns[(v == 1).any()]
            case.append(list(check))
        
    elif(((cat_columns == 1) and (total_null == 0)) or ((num_columns == 0) and (total_null == 0))):  
        case3 = pd.read_csv('pipelines/case3.csv')
        df_dict = dict(
            list(
                case3.groupby(case3.index)
            )
        )

        for k, v in df_dict.items():         
            check = v.columns[(v == 1).any()]
            case.append(list(check))
        
    elif(((cat_columns == 1) and (cat_null > 0)) or ((num_columns == 0) and (cat_null > 0))):
        case4 = pd.read_csv('pipelines/case4.csv')
        df_dict = dict(
            list(
                case4.groupby(case4.index)
            )
        )

        for k, v in df_dict.items():         
            check = v.columns[(v == 1).any()]
            case.append(list(check))
        
    else:
        case5 = pd.read_csv('pipelines/case5.csv')
        df_dict = dict(
            list(
                case5.groupby(case5.index)
            )
        )

        for k, v in df_dict.items():         
            check = v.columns[(v == 1).any()]
            case.append(list(check))
        
    return case
```

# Function for conducting the training

```python
def training(path, target, dataset):
    for i in range(len(pipeline)):
        print("Preprocess ", i+1, ": ", pipeline[i])
        data = cleaner()
        flag = 0
        
        df, y, _, _ = path.get_data(dataset_format="dataframe", target=target)
        
        if(isinstance(y,np.ndarray)):
            y = pd.DataFrame({target: y})

        df = pd.concat([df, y], axis=1)
        
        data.set_dataframe(df)
        
        if(i == 0):
            case = choose_case(data, target)
        
        if((pipeline[i] == []) or (pipeline[i] in case)):
            if(pipeline[i] != []):
                try:
                    data.preprocess(pipeline=pipeline[i], target=target)
                except:
                    flag = 2
                    print("Memory error... \n")
            if(flag == 0): 
                try:
                    df = data.get_dataframe()

                    # Training
                    X = df.drop([target], axis=1)
                    y = df[target]
                    X = X.to_numpy()
                    y = y.to_numpy()

                    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=1)

                    marvin_dataset = {'train_X': X_train, 'train_y': y_train, 'test_X': X_test, 'test_y': y_test}

                    rf = RandomForestClassifier(random_state=42, verbose=3)

                    try:
                        print("Start of the training...")
                        rf.fit(marvin_dataset['train_X'].copy(), marvin_dataset['train_y'].copy())

                        marvin_model = rf

                        # Metrics
                        predictions = marvin_model.predict(marvin_dataset['test_X'])
                        marvin_metrics = metrics.accuracy_score(marvin_dataset['test_y'], predictions)
                    except:
                        print("Training failed...")
                        marvin_metrics = 0
                except: 
                    print("Training failed...")
                    marvin_metrics = 0
        else:
            flag = 1
            marvin_metrics = 0

        if(flag == 1):
            print('Skip pipeline... \n')
        elif(flag == 2):
            marvin_metrics = 0
        else:
            print('Accuracy: ', marvin_metrics, '\n')

        dic['dataset_%s' % dataset].append((i, pipeline[i], marvin_metrics))
```

```python
datasets_openml = []

#15, 29, 179, 188, 443, 452, 455, 473, 575, 802, 839, 897, 930, 966, 990, 1024, 1037, 1053, 
#1119, 1205, 1351, 1354, 1358, 1364, 1367, 1369, 1373, 1375, 1402, 1403, 41526, 41430,
#1453, 1459, 1460, 1468, 1471, 1475, 1479, 1485, 1486, 1489, 1497, 1502, 1503, 1507, 1510, 
#1525, 1526, 1547, 1549, 1552, 1553, 1557, 1558, 1560, 1568, 1590, 4534, 4538, 4541, 6332, 
#23380, 23512, 23517, 40536, 40966, 40979, 40981, 40982, 40996, 40999, 41001, 41002, 41003, 
#41004, 41005, 41007, 41027, 41160, 42345, 42638
```

```python
import openml

for step in datasets_openml:
    dataset = openml.datasets.get_dataset(step)
    print("Dataset", step, ": ", dataset.url)
    try:
        with open('results.pickle', 'rb') as handle:
            dic = pickle.load(handle)
            dic['dataset_%s' % step] = []
    except:
        dic['dataset_%s' % step] = []
    training(dataset, dataset.default_target_attribute, step)
    with open('results.pickle', 'wb') as handle:
        pickle.dump(dic, handle, protocol=pickle.HIGHEST_PROTOCOL)
    print("======================= END OF THE DATASET ", step, " ======================= \n")
```
