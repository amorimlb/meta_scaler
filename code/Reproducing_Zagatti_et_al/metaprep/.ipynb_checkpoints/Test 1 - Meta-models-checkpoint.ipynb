# Test with Titanic

### 5 meta-models

```python
from metaprep import cleaner
import time
start_time = time.time()

data = cleaner()
data.load_csv('csv_results/titanic.csv')
results = data.meta_prep('Survived', verbose=1)
data.preprocess(pipeline=results[0], target='Survived', verbose=1)
data.test(algorithm='rf', target='Survived', verbose=1)

print("--- %s seconds ---" % (time.time() - start_time))
```

### Unique meta-model

```python
from metaprep import cleaner
import time
start_time = time.time()

data = cleaner()
data.load_csv('csv_results/titanic.csv')
results = data.unique_prep('Survived')
data.preprocess(pipeline=results[0], target='Survived', verbose=1)
data.test(algorithm='rf', target='Survived', verbose=1)

print("--- %s seconds ---" % (time.time() - start_time))
```
